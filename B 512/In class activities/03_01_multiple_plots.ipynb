Using `Patchwork`, can you re-create this plot?

![](img/patchwork_example.png)

```r
# uncomment to load the libraries

library("tidyverse")
library("patchwork")
options(repr.plot.res = 300)
```

👇use these cells to 👀 the data.

```r
iris %>% head
```

```r
mpg %>% head
```

```r
options(repr.plot.width= 9, repr.plot.height=4, repr.plot.res=300)
p= ggplot(mpg, aes(x=displ,y=hwy))
p= p+ geom_point()

q=ggplot(iris, aes(x=Sepal.Length,y=Sepal.Width,color=Species))
q= q+ geom_point()

p+q
```

What about this plot? 👇 What does this plot demonstrate?

![](img/patchwork_example_penguins.png)

```r
library("palmerpenguins")
penguins %>% head
```

```r
options(repr.plot.width= 9, repr.plot.height=4, repr.plot.res=300)
p=ggplot(penguins, aes(x=body_mass_g,y=flipper_length_mm,color=species))
p=p+ geom_point()

q=ggplot(penguins, aes(x=body_mass_g,y=flipper_length_mm,color=island))
q=q+ geom_point()+scale_color_brewer(type = "qual", palette = "Set2")

p+q
```

📍Hint: add `scale_color_brewer(type = "qual", palette = "Set2")` to adjust the colors of the second plot!
