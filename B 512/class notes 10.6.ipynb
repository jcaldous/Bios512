```r
options(repr.plot.width=7, rep.plot.height=3)
p=ggplot(mpg, aes(x=displ, y=hwy, color=class))
p=p+geom_point()
p
```

```r
p= p+ labs(x="Displacement, mL", y="Highway, MPG", color="vehicle\nclass")
p
```

```r
p=p+ theme(axis.title=element_text(color='red'))
p
```

```r
p=p+theme(plot.background=element_rect(fill='grey90'))
p
```

```r
p=p+theme(panel.background=element_rect(fill='light blue'))
p
```

```r
options(repr.plot.width=7, rep.plot.height=3)
q=ggplot(mpg, aes(x=displ, y=hwy, color=class))
q=q+geom_point()
q=q+labs(x="Displacement, mL", y="Highway, MPG", color="vehicle\nclass")
q=q+theme(
    axis.ticks.x= element_line(color='red'),
    plot.background=element_rect(fill='lightblue'),    
    plot.margin = margin(50,50,50,50),#top right bottom left
    axis.title.x= element_text(margin=margin(t=20))
    )
q
```
