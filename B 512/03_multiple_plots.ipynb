5. Create a facets plot where each facet represents one iris species. 
(You may have to resize the output again.)

```r
options(repr.plot.width=15, repr.plot.height=10)
p= ggplot(iris, aes(x=Petal.Width, y=Petal.Length, size=Sepal.Width, color=Species))
p=p+geom_point()
p= p+facet_grid(~Species)

p
```

6. Using `Patchwork`, can you re-create this plot?


![](img/patchwork_example.png)

```r
library('patchwork')
options(repr.plot.width=16, repr.plot.height=7)
p = ggplot(mpg, aes(x=displ, y=hwy))
p= p+geom_point(size=3)

q= ggplot(iris, aes(x=Sepal.Width, y=Sepal.Length, color= Species))
q=q+geom_point(size=3)

p+q
```
